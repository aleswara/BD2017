```python
#Loading Selenium Webdriver 
driver= webdriver.Chrome()
wait = WebDriverWait(driver, 5)
```

```python
driver.get('https://www.trekbikes.com/de/de_DE/store-finder/retailer/SE/')
#driver.get('https://www.trekbikes.com/de/de_DE/store-finder/retailer/SE/1/')
#driver.get('https://www.trekbikes.com/de/de_DE/store-finder/retailer/SE/2/')
time.sleep(3)
```

```python
#cookies
button=driver.find_element(By.ID, 'CybotCookiebotDialogBodyButtonAccept')
driver.execute_script("arguments[0].click();", button)
time.sleep(3)
```

```python
html = driver.page_source
soup = BeautifulSoup(html, "html.parser")
results= soup.find_all('div', {'class': 'vertical-spacer-4'})
```

```python
for result in results:
    name = [i.text[1:].split(",")[0] for i in result.find_all('span', {'class': 'text-strong'})]
    address = [i.text[1:].split(",")[0] for i in result.find_all('span', {'class': 'text-weak'})]
```

```python
print(name, address)
```

```python
# Do something with the parsed page, for example, print the title
#df = pd.DataFrame(columns={'Name': 'name','Address':'address'})
df = pd.DataFrame(address,name)
#df.set_axis(['Name'],axis=1)
#df = pd.DataFrame(data)
```
